## Deep Neural Network

### Initialization

```python
# Shape of the network
xn = X.shape[0]
yn = Y.shape[0]

# Number of units per layer
n = [xn, 4, 4, yn]
# Activations per layer
activations = ["relu", "relu", "softmax"]
```

```python
# Number of train examples
m = X.shape[1]
m
```

```python
def initialize_parameters(n):
    l = len(n) - 1
    params = {}
    np.random.seed(1)
    for i in range(l):
        params["W"+str(i+1)] = np.random.randn(n[i+1], n[i]) * 0.01
        params["b"+str(i+1)] = np.zeros((n[i+1], 1))

    return params
```

```python
initialize_parameters(n)
```

### Forward pass

```python
def forward_pass_relu(Z):
    return np.maximum(Z, 0.)
```

```python
relu_Z = np.array([[1,-2,3], 
                   [4,5,-6]])
forward_pass_relu(relu_Z)
```

```python
def forward_pass_softmax(Z):
    Z_exp = np.exp(Z)
    return Z_exp / np.sum(Z_exp, axis=0, keepdims=True)
```

```python
softmax_Z = np.array([[1,1,1], 
                      [2,1,1]])
forward_pass_softmax(softmax_Z)
#np.sum(softmax_Z, axis=0, keepdims=True)
```

```python
def forward_pass_activation(activation, Z):
    match activation:
        case "relu":
            return forward_pass_relu(Z)
        case "softmax":
            return forward_pass_softmax(Z)
        case _:
            raise Exception("No such activation as: "+activation)
```

```python
def forward_pass(X, n, activations, params):
    l = len(n) - 1
    A = X
    cache = {}
    cache["A"+str(0)] = A
    for i in range(l):
        W = params["W"+str(i+1)]
        b = params["b"+str(i+1)]
        Z = np.dot(W, A) + b
        A = forward_pass_activation(activations[i], Z)
        cache["Z"+str(i+1)] = Z
        cache["A"+str(i+1)] = A
        cache["W"+str(i+1)] = W

    return A, cache
```

```python
params = initialize_parameters(n)
A, cache = forward_pass(X, n, activations, params)
A.shape
```

### Calculating cost

```python
def cost(A, Y, m):
    J = - 1/m * np.sum(np.multiply(Y, np.log(A)))
    return J    
```

### Backward pass

```python
def backward_pass_output_layer_softmax(Y, l, m, cache, grads):
    #gradient from non-linear part
    A = cache["A"+str(l)]
    
    dZ = A - Y

    
    #gradient from linear part
    W = cache["W"+str(l)]
    A_prev = cache["A"+str(l-1)]
    
    grads["dW"+str(l)] = 1/m * np.dot(dZ, A_prev.T)
    grads["db"+str(l)] = 1/m * np.sum(dZ, axis= 1, keepdims=True)

    #retruns dA for the previous layer
    return np.dot(W.T, dZ)
```

```python
def backward_pass_output_layer(Y, n_output, l, activation, m, cache, grads):
    match activation:
        case "softmax":
            return backward_pass_output_layer_softmax(Y, l, m, cache, grads)
        case _:
            raise Exception("No such activation as: "+activation)
```

```python
def backward_pass_relu(Z):
    return np.where(Z > 0, 1, 0)
```

```python
def backward_pass_activation(activation, Z):
    match activation:
        case "relu":
            return backward_pass_relu(Z)
        case _:
            raise Exception("No such activation as: "+activation)
```

```python
def backward_pass_hidden_layers(dA_last, n, activations, m, params, grads):
    l = len(n) - 1
    dA = dA_last
    for i in reversed(range(l)):
        #gradient from non-linear part
        Z = cache["Z"+str(i+1)]
        
        dZ = dA * backward_pass_activation(activations[i], Z)

        
        #gradient from linear part
        W = cache["W"+str(i+1)]           
        A_prev = cache["A"+str(i)]
        
        grads["dW"+str(i+1)] = 1/m * np.dot(dZ, A_prev.T)
        grads["db"+str(i+1)] = 1/m * np.sum(dZ, axis = 1, keepdims=True)

        #dA for the previous layer
        dA = np.dot(W.T, dZ)

    return grads
```

```python
def backward_pass(Y, n, activations, m, cache):
    l = len(n) - 1
    grads = {}
    dA = backward_pass_output_layer(Y, n[-1], l, activations[-1], m, cache, grads)
    grads = backward_pass_hidden_layers(dA, n[:-1], activations[:-1], m, params, grads)
    return grads
```
